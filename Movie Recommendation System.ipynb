```python
df = df.drop_duplicates().reset_index(drop=True)
```

```python
df = df[['title', 'overview', 'genres', 'tagline', 'vote_average', 'popularity']]
```

```python
df.isnull().sum()
```

```python
df = df.dropna(subset=['title'])
```

```python
df['overview'] = df['overview'].fillna('')
```

```python
import ast
df['genres'] = df['genres'].apply(lambda x: " ".join([i['name'] for i in ast.literal_eval(x)]))
```

```python
df['tagline'] = df['tagline'].fillna('')
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
df['vote_average'] = df['vote_average'].fillna(df['vote_average'].mean())
```

```python
df['popularity'] = df['popularity'].fillna(df['popularity'].mode())
```

```python
df.isna().sum()
```

```python
df['tags'] = df['overview']+df['genres']+df['tagline']
```

```python
df.head()
```

```python
df = df.reset_index(drop=True)
```

```python
indices = pd.Series(df.index,index = df['title']).drop_duplicates()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(max_features=50000,ngram_range=(1,2),stop_words='english')
```

```python
tfidf_matrix = tfidf.fit_transform(df['tags'])
```

```python
tfidf_matrix.shape
```

```python
# Apply string methods to Series using .str accessor
# Then tokenize each row individually using apply()
df['tags'] = df['tags'].str.lower().apply(nltk.word_tokenize)

# Alternative approach if you want to tokenize the entire column as one string:
# df['tags'] = nltk.word_tokenize(' '.join(df['tags'].str.lower()))
```

```python
stop_words = set(stopwords.words('english'))
```

```python
df['tags'] = df['tags'].apply(lambda words: [w for w in words if w not in stop_words])
```

```python
# similarity = linear_kernel(tfidf_matrix, tfidf_matrix) this will kill

from sklearn.metrics.pairwise import linear_kernel

from sklearn.metrics.pairwise import linear_kernel

def recommend(title, n=10):
    if title not in indices:
        return ["Movie not found"]

    idx = indices[title]

    # Compute similarity only for this movie → SAFE!
    sim_scores = linear_kernel(tfidf_matrix[idx], tfidf_matrix).flatten()

    # Sort by score
    similar_indices = sim_scores.argsort()[::-1][1:n+1]

    return df['title'].iloc[similar_indices].tolist()
```

```python
recommend("Comic Book Confidential", 10)
```

```python
recommend('Avatar')
```
